```python
# Volvemos a cargar el dataset pero ignorando las filas irrelevantes
df = pd.read_excel('../data/Finanzas_Datos.xlsx',skiprows=2)

#Renombramos colunas para sean más entendibles 
df.columns = [
    'ID','Producto','Cliente','Col1','Año','Periodo','Cantidad','Ingresos',
    'Conteo','Col2','Col3','Costos','Margen','Col4','Pagina'
]

df.head(10)
```

```python
# Verificamos los datos nulos
df.isnull().sum()
```

✅ Paso 2: Limpieza de datos

```python
# Eliminar columnas irrelevantes
df.drop(columns=["Col1", "Col2", "Col3", "Col4", "Pagina"], inplace=True)
```

```python
# Elimianr filas completamente vacías
df.dropna(how="all", inplace=True)
```

```python
# Elimianr filas con valores nulos en columas clave:
df.dropna(subset=["Producto", "Cliente", "Año", "Periodo", "Cantidad", "Ingresos"], inplace=True)
```

```python
# Varificamos los datos
df.isnull().sum()
df.head(10)
```

1. Se eliminaron correctamente las columnas innecesarias.
2. No hay valores nulos en columnas críticas (Producto, Cliente, Año, Periodo, Cantidad, Ingresos, Costos).
3. La columna "Margen" aún tiene valores nulos esto puede deberse a que estos valores sean calculables (Ingresos - Costos). (En caso de que no, se eliminaran)

🔍 Paso 3: Manejo de valores nulos en "Margen"

```python
# Llenado de con datos (ingresos - costos)
df["Margen"] = df["Margen"].fillna(df["Ingresos"] - df["Costos"])
```

```python
# Verificar si sigue habiendo valores nulos
df.isnull().sum()
```

```python
# Identificar valores atípicos usando el rango intercuartílico (IQR)
def detectar_outliers(columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < lim_inf) | (df[columna] > lim_sup)]
    return outliers

outliers_ingresos = detectar_outliers('Ingresos')
outliers_costos = detectar_outliers('Costos')

print(f"Se encontraron {len(outliers_ingresos)} valores atípicos en Ingresos")
print(f"Se encontraron {len(outliers_costos)} valores atípicos en Costos")
```
