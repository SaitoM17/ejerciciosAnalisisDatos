```python
import pandas as pd
import numpy as np
```

```python
nombres = ['said','jared','misra','sam','simba']

lista_nombres = [np.random.choice(nombres) for i in range(20)]
lista_num = [np.random.choice(np.arange(20,50)) for i in range(20)]
lista_can = [np.random.choice(np.arange(11)) for i in range(20)]

nombres = ['said','jared','misra','sam','simba']

lista_nombres_2 = [np.random.choice(nombres) for i in range(40)]
lista_num_2 = [np.random.choice(np.arange(15,60)) for i in range(40)]
lista_can_2 = [np.random.choice(np.arange(17)) for i in range(40)]

dic_1 = {
    'nombre': lista_nombres,
    'cantidad': lista_can,
    'precio': lista_num
}

dic_2 = {
    'nombre': lista_nombres_2,
    'cantidad': lista_can_2,
    'precio': lista_num_2
}


df_1 = pd.DataFrame(dic_1)
df_2 = pd.DataFrame(dic_2)
```
