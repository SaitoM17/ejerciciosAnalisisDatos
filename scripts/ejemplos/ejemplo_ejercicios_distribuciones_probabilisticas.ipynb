📌 Ejercicio 1: Simulación de Lanzamientos de Moneda (Distribución Binomial)

💡 Objetivo: Simular el lanzamiento de una moneda 1000 veces y analizar la distribución de los resultados.

🔹 Instrucciones:
1. Simula 1000 experimentos donde lanzas una moneda 10 veces en cada experimento.
2. Usa la distribución binomial (binom.rvs) con n=10 intentos y p=0.5.
3. Grafica un histograma con los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Parámetros
n, p = 10, 0.5  # 10 intentos, 50% de probabilidad de éxito
experimentos = 1000  

# Simulación
resultados = binom.rvs(n, p, size=experimentos)

# Visualización
plt.figure(figsize=(8, 5))
sns.histplot(resultados, bins=n+1, kde=False, color="blue")
plt.xlabel("Número de caras en 10 lanzamientos")
plt.ylabel("Frecuencia")
plt.title("Distribución Binomial: Lanzamientos de Moneda")
plt.show()
```

📌 Ejercicio 2: Análisis de Llegadas de Clientes (Distribución de Poisson)

💡 Objetivo: Simular cuántos clientes llegan a una tienda cada minuto.

🔹 Instrucciones:
1. Supón que la tienda recibe en promedio 5 clientes por minuto (λ=5).
2. Genera 1000 simulaciones usando la distribución de Poisson (poisson.rvs).
3. Grafica los resultados en un histograma.

```python
from scipy.stats import poisson

# Parámetro: tasa promedio de llegadas
lambda_poisson = 5  
experimentos = 1000  

# Simulación
clientes_por_minuto = poisson.rvs(lambda_poisson, size=experimentos)

# Visualización
plt.figure(figsize=(8, 5))
sns.histplot(clientes_por_minuto, bins=15, kde=False, color="red")
plt.xlabel("Clientes por minuto")
plt.ylabel("Frecuencia")
plt.title("Distribución de Poisson: Llegadas de Clientes")
plt.show()
```

✅ Preguntas para reflexionar:
1. ¿Cuál es el número de clientes más frecuente por minuto?
R= 160 por 4 minutos
2. ¿Cómo cambiaría la gráfica si la tienda recibiera en promedio 8 clientes por minuto?
R = Como se muestra en la imagen si son 8 cliente por minuto la frecuencia seria de 250 y tomaria una forma "casi" de campana.

📌 Ejercicio 3: Alturas de Personas (Distribución Normal)

💡 Objetivo: Modelar la distribución de alturas de una población.

🔹 Instrucciones:
1. Supón que la altura promedio de los adultos es 1.70 metros, con una desviación estándar de 0.10 metros.
2. Genera 1000 alturas simuladas con la distribución normal (np.random.normal).
3. Grafica un histograma con KDE para visualizar la distribución.

```python
# Parámetros
media_altura = 1.70  
desviacion_altura = 0.10  
num_muestras = 1000  

# Simulación
alturas = np.random.normal(media_altura, desviacion_altura, num_muestras)

# Visualización
plt.figure(figsize=(8, 5))
sns.histplot(alturas, bins=30, kde=True, color="purple")
plt.xlabel("Altura (m)")
plt.ylabel("Frecuencia")
plt.title("Distribución Normal: Alturas de Personas")
plt.show()
```

✅ Preguntas para reflexionar:
1. ¿Cómo se ve la distribución? ¿Parece una campana?
R = Si, la distribución parace una campana.
2. ¿Cómo cambiaría la gráfica si la desviación estándar fuera mayor (ejemplo: 0.20 en vez de 0.10)?
R = Como tan no cambiaria mucho a simple vista, pero en realidad parece tener una leve inclinación al lado derecho.
