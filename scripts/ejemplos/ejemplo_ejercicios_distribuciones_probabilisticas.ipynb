üìå Ejercicio 1: Simulaci√≥n de Lanzamientos de Moneda (Distribuci√≥n Binomial)

üí° Objetivo: Simular el lanzamiento de una moneda 1000 veces y analizar la distribuci√≥n de los resultados.

üîπ Instrucciones:
1. Simula 1000 experimentos donde lanzas una moneda 10 veces en cada experimento.
2. Usa la distribuci√≥n binomial (binom.rvs) con n=10 intentos y p=0.5.
3. Grafica un histograma con los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Par√°metros
n, p = 10, 0.5  # 10 intentos, 50% de probabilidad de √©xito
experimentos = 1000  

# Simulaci√≥n
resultados = binom.rvs(n, p, size=experimentos)

# Visualizaci√≥n
plt.figure(figsize=(8, 5))
sns.histplot(resultados, bins=n+1, kde=False, color="blue")
plt.xlabel("N√∫mero de caras en 10 lanzamientos")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n Binomial: Lanzamientos de Moneda")
plt.show()
```

R= La grafica cambia se inclina m√°s para el lado derecho

üìå Ejercicio 2: An√°lisis de Llegadas de Clientes (Distribuci√≥n de Poisson)

üí° Objetivo: Simular cu√°ntos clientes llegan a una tienda cada minuto.

üîπ Instrucciones:
1. Sup√≥n que la tienda recibe en promedio 5 clientes por minuto (Œª=5).
2. Genera 1000 simulaciones usando la distribuci√≥n de Poisson (poisson.rvs).
3. Grafica los resultados en un histograma.

```python
from scipy.stats import poisson

# Par√°metro: tasa promedio de llegadas
lambda_poisson = 5  
experimentos = 1000  

# Simulaci√≥n
clientes_por_minuto = poisson.rvs(lambda_poisson, size=experimentos)

# Visualizaci√≥n
plt.figure(figsize=(8, 5))
sns.histplot(clientes_por_minuto, bins=15, kde=False, color="red")
plt.xlabel("Clientes por minuto")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de Poisson: Llegadas de Clientes")
plt.show()
```

‚úÖ Preguntas para reflexionar:
1. ¬øCu√°l es el n√∫mero de clientes m√°s frecuente por minuto?
R= 160 por 4 minutos
2. ¬øC√≥mo cambiar√≠a la gr√°fica si la tienda recibiera en promedio 8 clientes por minuto?
R = Como se muestra en la imagen si son 8 cliente por minuto la frecuencia seria de 250 y tomaria una forma "casi" de campana.

üìå Ejercicio 3: Alturas de Personas (Distribuci√≥n Normal)

üí° Objetivo: Modelar la distribuci√≥n de alturas de una poblaci√≥n.

üîπ Instrucciones:
1. Sup√≥n que la altura promedio de los adultos es 1.70 metros, con una desviaci√≥n est√°ndar de 0.10 metros.
2. Genera 1000 alturas simuladas con la distribuci√≥n normal (np.random.normal).
3. Grafica un histograma con KDE para visualizar la distribuci√≥n.

```python
# Par√°metros
media_altura = 1.70  
desviacion_altura = 0.10  
num_muestras = 1000  

# Simulaci√≥n
alturas = np.random.normal(media_altura, desviacion_altura, num_muestras)

# Visualizaci√≥n
plt.figure(figsize=(8, 5))
sns.histplot(alturas, bins=30, kde=True, color="purple")
plt.xlabel("Altura (m)")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n Normal: Alturas de Personas")
plt.show()
```

‚úÖ Preguntas para reflexionar:
1. ¬øC√≥mo se ve la distribuci√≥n? ¬øParece una campana?
R = Si, la distribuci√≥n parace una campana.
2. ¬øC√≥mo cambiar√≠a la gr√°fica si la desviaci√≥n est√°ndar fuera mayor (ejemplo: 0.20 en vez de 0.10)?
R = Como tan no cambiaria mucho a simple vista, pero en realidad parece tener una leve inclinaci√≥n al lado derecho.
