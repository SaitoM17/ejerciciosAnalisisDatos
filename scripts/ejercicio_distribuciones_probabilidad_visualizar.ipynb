游늷 Ejercicio: Visualizar Distribuciones de Probabilidad

Generaremos 4 distribuciones y las graficaremos:
1. Uniforme
2. Binomial
3. Poisson
4. Normal

C칩digo en Python
- Este c칩digo generar치 datos simulados para cada distribuci칩n y los visualizar치 con histogramas y curvas de densidad.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm, uniform

# Configuraci칩n de la figura
plt.figure(figsize=(12, 8))

# 1勇 Distribuci칩n Uniforme
plt.subplot(2, 2, 1)
datos_uniforme = np.random.uniform(0, 10, 1000)  # N칰meros aleatorios entre 0 y 10
sns.histplot(datos_uniforme, bins=30, kde=True, color="blue")
plt.title("Distribuci칩n Uniforme")

# 2勇 Distribuci칩n Binomial
plt.subplot(2, 2, 2)
n, p = 10, 0.5  # 10 intentos, probabilidad de 칠xito 50%
datos_binomial = binom.rvs(n, p, size=1000)
sns.histplot(datos_binomial, bins=10, kde=False, color="green")
plt.title("Distribuci칩n Binomial")

# 3勇 Distribuci칩n de Poisson
plt.subplot(2, 2, 3)
lambda_poisson = 5  # Media esperada de ocurrencias
datos_poisson = poisson.rvs(lambda_poisson, size=1000)
sns.histplot(datos_poisson, bins=15, kde=False, color="red")
plt.title("Distribuci칩n de Poisson")

# 4勇 Distribuci칩n Normal
plt.subplot(2, 2, 4)
media, desviacion = 0, 1  # Media = 0, Desviaci칩n est치ndar = 1
datos_normal = np.random.normal(media, desviacion, 1000)
sns.histplot(datos_normal, bins=30, kde=True, color="purple")
plt.title("Distribuci칩n Normal")

plt.tight_layout()
plt.show()
```
