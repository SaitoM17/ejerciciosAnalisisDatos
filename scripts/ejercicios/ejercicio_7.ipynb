游늷 Matplotlib
Matplotlib es la biblioteca principal de visualizaci칩n en Python. Proporciona una gran flexibilidad para crear gr치ficos de distintos tipos, como gr치ficos de l칤neas, dispersi칩n, histogramas, barras, etc.

Elementos clave de Matplotlib
1. pyplot: Es un m칩dulo de Matplotlib que facilita la creaci칩n de gr치ficos con una sintaxis similar a MATLAB.
2. Figura (Figure): Es el contenedor principal de la visualizaci칩n.
3. Ejes (Axes): Son los sistemas de coordenadas dentro de una figura donde se colocan los gr치ficos.
4. T칤tulos y etiquetas: Se pueden agregar t칤tulos y etiquetas a los ejes con plt.title(), plt.xlabel(), plt.ylabel().
5. Leyendas: Se usa plt.legend() para mostrar etiquetas en los gr치ficos.

Ejemplo b치sico de Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear la figura y los ejers
plt.figure(figsize=(8,5))

# Crear el gr치fico
plt.plot(x, y, label='seno', color='b', linestyle='--')

# A침adir titulo
plt.title('Gr치fico de la funci칩n seno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.legend()
plt.grid(True)

# Mostrar gr치fico
plt.show()
```

游늷 Seaborn
Seaborn es una biblioteca basada en Matplotlib que simplifica la visualizaci칩n de datos y a침ade funcionalidades avanzadas.

Caracter칤sticas principales
1. Estilo mejorado: Seaborn ofrece gr치ficos m치s atractivos con menos configuraci칩n.
2. Facilidad de uso: Permite crear gr치ficos complejos con menos c칩digo.
3. Soporte para DataFrames de Pandas: Se integra bien con Pandas para graficar directamente desde DataFrames.
4. Gr치ficos estad칤sticos: Incluye gr치ficos como histogramas, boxplots y gr치ficos de regresi칩n.

Ejemplo b치sico de Seaborn

```python
import pandas as pd
import seaborn as sns

# Crear un DatFrame de ejemplo
np.random.seed(42)
data = pd.DataFrame({
    'Categoria': ['A', 'B', 'C', 'D', 'E'],
    'Valor 1': np.random.randn(5),
})

# Configurar estilo
sns.set_style('whitegrid')

# Crear un boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='Categoria', y='Valor 1', data=data, palette='coolwarm')

# Agregar tittulo
plt.title('Dsitribuci칩n de la variable por categor칤a')

#Mostrar gr치fico
plt.show()
```
