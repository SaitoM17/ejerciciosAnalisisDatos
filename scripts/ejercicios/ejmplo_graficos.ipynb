```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
ventas = [12000, 15000, 17000, 16000, 18000, 20000]
productos_vendidos = [300, 350, 400, 380, 420, 450]
clientes_nuevos = [50, 65, 70, 60, 80, 90]
tiempo_entrega = np.random.normal(loc=5, scale=1.2, size=100)  # Simulando tiempos de entrega
```

```python
# Configurar estilo
sns.set_theme(style="whitegrid")

# 1. Gráfico de líneas (Evolución de ventas)
plt.figure(figsize=(8, 5))
plt.plot(meses, ventas, marker='o', linestyle='-', color='b', label='Ventas')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.title('Evolución de Ventas en 6 meses')
plt.legend()
plt.show()
```

```python
# 2. Gráfico de barras (Comparación de productos vendidos por mes)
plt.figure(figsize=(8, 5))
sns.barplot(x=meses, y=productos_vendidos, palette='viridis')
plt.xlabel('Mes')
plt.ylabel('Productos Vendidos')
plt.title('Cantidad de Productos Vendidos por Mes')
plt.show()
```

```python
# 3. Gráfico de dispersión (Clientes nuevos vs Ventas Totales)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=clientes_nuevos, y=ventas, color='r')
plt.xlabel('Clientes Nuevos')
plt.ylabel('Ventas Totales ($)')
plt.title('Relación entre Clientes Nuevos y Ventas')
plt.show()
```

```python
# 4. Histograma (Distribución del tiempo de entrega promedio)
plt.figure(figsize=(8, 5))
sns.histplot(tiempo_entrega, bins=15, kde=True, color='g')
plt.xlabel('Tiempo de Entrega (días)')
plt.ylabel('Frecuencia')
plt.title('Distribución del Tiempo de Entrega Promedio')
plt.show()
```

```python
# Extra: Gráfico de áreas apiladas (Ventas por mes en proporción)
plt.figure(figsize=(8, 5))
plt.stackplot(meses, ventas, colors=['blue'], alpha=0.5)
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.title('Proporción de Ventas por Mes')
plt.show()
```
