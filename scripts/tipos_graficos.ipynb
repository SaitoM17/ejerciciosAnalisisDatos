```python
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas = [100, 120, 130, 110, 150, 170]

plt.plot(meses, ventas, marker='o', linestyle='-', color='b')
plt.title('Ventas Mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    "Sucursal": ["A", "B", "C"],
    "Ventas": [300, 450, 500]
})

sns.barplot(x="Sucursal", y="Ventas", data=data, palette="Blues")
plt.title("Ventas por Sucursal")
plt.show()
```

```python
import numpy as np

# Datos simulados
np.random.seed(42)
publicidad = np.random.rand(50) * 1000
ventas = publicidad * 0.5 + np.random.randn(50) * 50

plt.scatter(publicidad, ventas, color='r')
plt.title("Relacin entre Publicidad y Ventas")
plt.xlabel("Publicidad ($)")
plt.ylabel("Ventas ($)")
plt.show()
```

```python
edades = np.random.normal(35, 10, 1000)  # Media de 35 aos, desviacin estndar de 10

sns.histplot(edades, bins=20, kde=True, color='g')
plt.title("Distribucin de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
data = pd.DataFrame({
    "Departamento": ["Ventas"] * 50 + ["IT"] * 50 + ["RRHH"] * 50,
    "Salario": np.concatenate([np.random.normal(50000, 5000, 50),
                               np.random.normal(60000, 6000, 50),
                               np.random.normal(45000, 4000, 50)])
})

plt.figure(figsize=(8,5))
sns.boxplot(x="Departamento", y="Salario", data=data, palette="coolwarm")
plt.title("Distribucin de Salarios por Departamento")
plt.show()

```

```python
# Crear datos de correlacin
data = pd.DataFrame(np.random.rand(10, 10), columns=list('ABCDEFGHIJ'))

plt.figure(figsize=(8,6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

```python
# Generar datos de ventas de dos productos
np.random.seed(42)
dias = np.arange(1, 31)
ventas_a = np.cumsum(np.random.randint(10, 50, size=30))
ventas_b = np.cumsum(np.random.randint(20, 60, size=30))

# Crear DataFrame
data = pd.DataFrame({"Da": dias, "Producto A": ventas_a, "Producto B": ventas_b})

# Graficar con lneas suavizadas
plt.figure(figsize=(10, 5))
sns.lineplot(x="Da", y="Producto A", data=data, label="Producto A", marker='o')
sns.lineplot(x="Da", y="Producto B", data=data, label="Producto B", marker='s')

# Resaltar un periodo de inters
plt.axvspan(10, 15, color='gray', alpha=0.2, label="Promocin Especial")

plt.title("Comparacin de Ventas de Productos")
plt.xlabel("Da")
plt.ylabel("Ventas Acumuladas")
plt.legend()
plt.show()
```

```python
# Datos de ventas de 3 productos en diferentes trimestres
trimestres = ["Q1", "Q2", "Q3", "Q4"]
ventas = {
    "Producto A": [200, 250, 300, 280],
    "Producto B": [180, 220, 270, 260],
    "Producto C": [210, 260, 310, 290],
}

df = pd.DataFrame(ventas, index=trimestres)

# Crear grfico de barras agrupadas
df.plot(kind='bar', figsize=(10, 5), colormap='coolwarm')
plt.title("Ventas Trimestrales por Producto")
plt.xlabel("Trimestre")
plt.ylabel("Ventas")
plt.xticks(rotation=0)
plt.legend(title="Productos")
plt.show()
```

```python
# Crear grfico de barras apiladas
df.plot(kind='bar', stacked=True, figsize=(10, 5), colormap='coolwarm')
plt.title("Ventas Trimestrales Acumuladas")
plt.xlabel("Trimestre")
plt.ylabel("Ventas")
plt.xticks(rotation=0)
plt.legend(title="Productos")
plt.show()
```

```python
# Generar datos de marketing y ventas
np.random.seed(42)
marketing = np.random.rand(100) * 10000
ventas = marketing * 0.5 + np.random.randn(100) * 500 + 2000

df = pd.DataFrame({"Marketing": marketing, "Ventas": ventas})

# Grfico de dispersin con lnea de regresin
plt.figure(figsize=(8, 5))
sns.regplot(x="Marketing", y="Ventas", data=df, scatter_kws={"alpha": 0.6}, line_kws={"color": "red"})
plt.title("Inversin en Marketing vs Ventas")
plt.xlabel("Inversin en Marketing ($)")
plt.ylabel("Ventas ($)")
plt.show()
```

```python
# Crear grfico de hexgonos
plt.figure(figsize=(8, 5))
plt.hexbin(marketing, ventas, gridsize=30, cmap='Blues', alpha=0.8)
plt.colorbar(label="Frecuencia")
plt.title("Densidad de Puntos - Marketing vs Ventas")
plt.xlabel("Inversin en Marketing ($)")
plt.ylabel("Ventas ($)")
plt.show()
```

```python
# Generar datos de salarios por industria
data = pd.DataFrame({
    "Industria": ["Finanzas"] * 50 + ["Tecnologa"] * 50 + ["Salud"] * 50,
    "Salario": np.concatenate([
        np.random.normal(70000, 8000, 50),
        np.random.normal(85000, 9000, 50),
        np.random.normal(60000, 7000, 50)
    ])
})

plt.figure(figsize=(8,5))
sns.boxplot(x="Industria", y="Salario", data=data, palette="viridis")
plt.title("Distribucin de Salarios por Industria")
plt.show()
```
